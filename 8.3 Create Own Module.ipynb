# Create own Module

A module is simply a Python file with a .py extension that can be imported inside another Python program. The name of the Python file becomes the module name. The module contains definitions and implementation of classes, variables, and functions that can be used inside another program.

Example: Let’s create a simple module named newModule

```python
# Python program to create
# a module named newModule  
  
# Defining a function
def myFunc():
    print("This is Module Function")
  
# Defining a variable
location = "India"
```

* The above example shows the creation of a simple module named newModule as the name of the above Python file is newModule.py.
* When this code is executed it does nothing because the function created is not invoked.

## How to use Module?
To use the above created module, create a new Python file in the same directory and import newModule module using the import statement.

```python
# Python program to demonstrate
# modules
import newModule

# Use the function created
newModule.myFunc()

# Print the variable declared
print(newModule.location)
```

The above example shows a simple module with only functions and variables. Now let’s create a little bit complex module with classes, functions, and variables. Below is the implementation.

Example: Open the above created newModule module and make the following changes.

```python
# Python program to create
# a module named newModule  
  
# Defining a function
def myFunc():
    print("This is Module Function")
  
# Defining a variable
location = "India"

# Defining a class
class Employee():
      
    def __init__(self, name, position):
        self. name = name
        self.position = position
          
    def show(self):
        print("Employee name:", self.name)
        print("Employee position:", self.position)
```

In this example, a class named employee has been declared with a method show() to print the details of the employee. Now open the Python script for importing and using this module.

Example:

```python
# Python program to demonstrate
# modules  
   
import newModule   
  
# Use the class created
emp = newModule.Employee("Sam", "Developer")
emp.show()
```

**If the imported Module is not present in this case the program gives error.**<br>
Let's see Example:

```python
import nonexistmodule  
print(nonexistmodule.newword)
```

**To Avoid this we use try statement with an except ImportError clause can be used to guard against unsuccessful import attempts:**

```python
try:
    # Non-existent module
    import nonexistmodule
    print(nonexistmodule.newword)
except ImportError:
    print('Object not found in module')
```
